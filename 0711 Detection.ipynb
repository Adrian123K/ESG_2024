```python
import pandas as pd
import os


# 이미지 폴더에서 파일 목록 가져오기
image_files = os.listdir('images_001')

# 임의의 박스 좌표 생성 (예시)
data = []
for img_file in image_files[:100]:  # 처음 100개 이미지만 사용
    data.append({
        'image_id': img_file,
        'x_min': 100,
        'y_min': 100,
        'x_max': 300,
        'y_max': 300,
        'class': 'abnormality'
    })

# DataFrame 생성 및 CSV 파일로 저장
df = pd.DataFrame(data)
df.to_csv('train.csv', index=False)
print("Created train.csv with", len(df), "entries.")
```
