#### 전역변수, 지역변수
    - 활성화된 세션내의 변수는 어디서든 사용가능
    - 지역변수: 함수 내부의 변수는 함수 안에서만 사용되고 외부로 뽑아낼 수 없음
    - 전역변수: 외부변수를 함수 내부에서도 사용하고 외부에도 그대로 반영할 수 있도록 선언
        + global 선언

````python
num = 702

def today():
    print(num)

today()
````

````python
num = 702

def tommorow():
    num = 703 # 지역변수
    print(f"tommorow: {num}")

tommorow()

print(f"today: {num}")
````

````python
num = 218

def birth():
    global num
    print(f"num: {num}")

    num = 621
    print(f"date: {num}")

birth()

print(f"num: {num}")
````

### 클래스
    - 파이썬에서 모듈화하여 사용하는 대표적인 형태
    - 클래스 내부에 변수, 함수 등을 선언하여 사용
    - 클래스 인스턴스화를 통해 객체지향 프로그래밍

````python
class Study:
    def __init__(self): # 생성자: 클래스 인스턴스 생성 시 자동으로 실행되는 함수
        self.ls = []
        self.student = []
        print(f"Study Class가 생성되었습니다.")

    def join_study(self, name, age, cls):
        self.student.append(name)
        self.student.append(int(age))
        self.student.append(cls)

        self.ls.append(self.student)

        print(f"스터디에 {self.ls[-1]}가 참여하였습니다.")
````

````python
ml_study = Study() # 클래스 인스턴스 생성
````

````python
ml_study.join_study("ASK", "20", "Math")
````

#### 클래스 상속
    - 부모 클래스의 내용을 그대로 받아 사용(상속)
    - 필요에 따라 부모 클래스의 함수나 변수를 변경하여 사용

````python
class GroupStudy(Study):
    def __init__(self):
        super().__init__()
        self.group_ls = []

    def group_est(self, rep):
        self.group_ls.append(rep)
        print(f"스터디 그룹 {self.group_ls[-1]}(이)가 생성되었습니다.")

        if self.ls:
            print("참여자 명단>")
            for p in self.ls:
                print("\t", p[0])
        else:
            print("참여자가 없습니다.")
````

````python
ml_gstudy = GroupStudy()
````

````python
ml_gstudy.group_est("Machine Learning")
````

````python
ml_study.group_est("ML")
````

````python
ml_gstudy.join_study("ASK", "20", "Math")
````

````python
ml_gstudy.group_est("Math")
````

### 예외처리
    - Try, Except 구문을 활용하여 예외처리 시행
    - 코드가 예외발생시 정상 종료가 되도록 설정하여 사용

````python
pw = input("비밀번호를 입력하세요: ")
try:
    print(f"비밀번호의 총 합은 {sum(map(int, pw.split(' ')))}입니다.")
except:
    print("올바른 비밀번호 타입이 아닙니다.")
````

````python
# try except else, try except finally
pw = input("비밀번호를 입력하세요: ")

try:
    res = sum(map(int, pw.split(" ")))
    print(f"비밀번호의 총 합은 {res}입니다.")
except:
    print("올바른 비밀번호 형식이 아닙니다.")
else: # 정상 실행이 되었음을 확인할 수 있다.
    print("이용해주셔서 감사")
````

````python
pw = input("비밀번호를 입력하세요: ")

try:
    res = sum(map(int, pw.split(" ")))
    # print(f"비밀번호의 총 합은 {res}입니다.")
except:
    print("올바른 비밀번호 형식이 아닙니다.")
else:
    if res < 40:
        print("총 합이 조건에 미치지 않으므로 실행할 수 없습니다.")
    else:
        print("Unlocked")
finally:
    print("이용해주셔서 감사합니다.")
````

#### 사용자 정의 예외처리
    - python에서는 예외가 아니나 사용자가 조건을 설정하여 예외를 발생시켜 비정상종료 시키는 방법
    - 오버라이딩 또는 새로운 클래스 선언을 통해 생성

````python
class LowNumber(Exception):
    def __str__(self):
        return "조건이 맞지 않습니다."

def pw_input():
    num = sum(map(int, input("비밀번호를 입력하세요: ").split(" ")))
    if num < 50:
        raise LowNumber()
    print(num)
````
